## Indian_retail_sales

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset
print(":-:-:-:-:-:-:-Load Dataset:-:-:-:-:-:-:-")
df = pd.read_csv('data.csv')
print(":-:-:-:-:-:-:-Display first 5 rows of the dataset:-:-:-:-:-:-:-")
df.head()
```

## Last data of file

```python
print(":-:-:-:-:-:-:-Display last 5 rows of the dataset:-:-:-:-:-:-:-")
df.tail()
```

## Information about dataset

```python
print(":-:-:-:-:-:-:-Dataset Information:-:-:-:-:-:-:-")
df.info()
```

## Description about dataset

```python
print(":-:-:-:-:-:-:-Statistical Summary of the Dataset:-:-:-:-:-:-:-")
df.describe()
```

## Shape of dataset

```python
print(":-:-:-:-:-:-:-Dataset Shape (Rows, Columns):-:-:-:-:-:-:-")
df.shape
```

## Read the columns

```python
print(":-:-:-:-:-:-:-Dataset Columns:-:-:-:-:-:-:-")
df.columns
```

## check the nulls

```python
print(":-:-:-:-:-:-:-Missing Values in Each Column:-:-:-:-:-:-:-")
df.isnull().sum()
df.head(11)
```

## Check the dtypes

```python
print(":-:-:-:-:-:-:-Bar chart for Category VS Revenue:-:-:-:-:-:-:-:-")
df.convert_dtypes()
```

## Fill the Null Values

```python
print(":-:-:-:-:-:-:-Handling Missing Values by Forward Fill Method:-:-:-:-:-:-:-")
print(df.isnull().sum())
df.empty
print("If any missing values are present show here.....")
```

## Category - Wise Performance

```python
print(":-:-:-:-:-:-:-Category wise performance:-:-:-:-:-:-:-")
category_performance = df.groupby('Product_Category')['Quantity'].sum().sort_values(ascending=False)
print(category_performance)
```

## Region Wise total revenue

```python
print(":-:-:-:-:-:-:-Region Wise Total Revenue:-:-:-:-:-:-:-")
region_revenue = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)
print(region_revenue)
```

## Drop Duplicates

```python
print(":-:-:-:-Drop The Duplicates:-:-:-:-")
df.drop_duplicates(inplace=True)
df
```

## Cleaned data

```python
print(":-:-:-:-:-:-:-Save the Cleaned Data to a New CSV File:-:-:-:-:-:-:-")
df.to_csv('data.csv')
print("Cleaned data saved to 'data.csv'.....")
```

## Customer type comparsion ( New VS Returning)

```python
print(":-:-:-:-:-:-:-Customer type comparison (New vs Returning):-:-:-:-:-:-:-:-")
customer_revenue = df.groupby('Customer_Type')['Revenue'].sum()

print(customer_revenue)
```

## Impact of discounts on revenue show we show real existence of discounts which affect sales or not

```python
print(":-:-:-:-:-:-:-Impact of Discounts on revenue:-:-:-:-:-:-:-")
plt.figure(figsize=(10,6))
sns.lineplot(df, x='Discount_Percent', y='Revenue', color='green')
plt.title('Impact of Discounts on Revenue')
plt.xlabel('Discount')
plt.ylabel('Revenue')
plt.show()
```

## Here with the Bar chart (Category VS Revenue ) comparsion show here

```python
print(":-:-:-:-:-:-:-Bar chart for Category VS Revenue:-:-:-:-:-:-:-:-")
category_revenue = df.groupby('Product_Category')['Revenue'].sum()
plt.figure(figsize=(10,6))
sns.barplot(x=category_revenue.index, y=category_revenue.values, color='purple')
plt.title('Category vs Revenue')
plt.xlabel('Category')
plt.ylabel('Total Revenue')
plt.show()
```

## Histogram (Revenue distribution and frequncy are shown)

```python
print(":-:-:-:-:-:-:-Histogram for Revenue Distribution:-:-:-:-:-:-:-")
plt.figure(figsize=(10,6))
sns.histplot(df['Revenue'], bins=30, kde=True, color='green')
plt.title('Distribution of Revenue')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()
```

## Pie Plot (Proportion of Categories which is shown by pie chart)

```python
print(":-:-:-:-:-:-:-Pie Plot (Propotion of categories):-:-:-:-:-:-:-")
category_counts = df['Product_Category'].value_counts()
plt.figure(figsize=(10,8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=150)
plt.title('Proportion of Product Categories')
plt.show()
```

## With the help of scatter plot here we see the Revenue VS Quantity Sold in this plot

```python
print(":-:-:-:-:-:-:-Scatter Plot (Revenue vs Quantity sold):-:-:-:-:-:-:-")
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Quantity', y='Revenue', hue='Product_Category')
plt.title('Revenue vs Quantity Sold')
plt.xlabel('Quantity Sold')
plt.ylabel('Revenue')
plt.show()
```
