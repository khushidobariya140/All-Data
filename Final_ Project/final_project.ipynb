##FINAL PROJECT TOPIC: Student Performance Analysis

##Load Dataset Of Student Performance Analysis

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load the dataset
print(":-:-:-:-Student Performance Analysis:-:-:-:- ")
df=pd.read_csv( "student_performance_dataset.csv" )
df.head()
```

##Shape of Dataset

```python
#here we show shape of dataset
print(":-:-:-:-Shape Of The Dataset:-:-:-:-")
df.shape
df.head()
```

##Convert datatype 

```python
#dtypes() used to convert datatypes ont to another
print(":-:-:-:-Convert Datatypes:-:-:-:-")
df = df.convert_dtypes()
df.head()
```

##Information about dataset

```python
#information of the our dataset
print(":-:-:-:-Information Of All Columns:-:-:-:-\n")
df.info()
```

##Description

```python
#This function used for statisctical description
print(":-:-:-:-Description:-:-:-:-\n")
df.describe()
```

##Null value

```python
#For Null values
print(":-:-:-:-For The Null Values:-:-:-:-\n")
print(df.isnull().sum())
df.head()
```

##Handle missing value

```python
#fillna,isnull used to handles missing values
print(":-:-:-:-Handle the null values:-:-:-:-\n")
df.fillna(df.mean(numeric_only=True), inplace=True)
df.isnull().sum()
```

#Cleaned data saved

```python
#Here saved cleaned data
print(":-:-:-:-Saved:-:-:-:-\n")
df.to_csv("cleaned_student_performance.csv", index=False)
print(" Cleaned data saved successfully.")
```

##Calculate total,percentage,grade

```python
#calculation applies on Subject,Total,Percentage or Grade
print(":-:-:-:-Calculations:-:-:-:-")
subjects = ['Maths','Science','English','History','Computer']
df['Total'] = df[subjects].sum(axis=1)
df['Percentage'] = round(df['Total']/len(subjects),2)

def assign_grade(g):
    if g>=90: return 'A+'
    elif g>=80: return 'A'
    elif g>=70: return 'B'
    elif g>=60: return 'C'
    elif g>=50: return 'D'
    else: return 'F'

df['Grade'] = df['Percentage'].apply(assign_grade)
df.head()
```

##Drop Duplicate values

```python
#Here drops duplicate values
print(":-:-:-:-Drop The Duplicates:-:-:-:-")
df.drop_duplicates(inplace=True)
df
```

##Search Student

```python
#search the specific student
print(":-:-:-:-Searching:-:-:-:-")
search_name = input("Enter student name to search: ")
result = df[df['Name'].str.contains(search_name, case=False, na=False)]
if not result.empty:
    print("\n Search Result:")
    print(result)
else:
    print(" No student found with that name.")
```

##Sort student by percentage

```python
#Sort student by Percentage
print(":-:-:-:-Sorting By Percentage:-:-:-:-")
sorted_df = df.sort_values(by='Percentage', ascending=False)
print("\nTop 10 Students by Percentage:")
print(sorted_df.head(10))
```

##Topper and Fail Students

```python
#List od Topper and Fail Students
print(":-:-:-:-Topper and Fail Students Detail:-:-:-:-")
topper = df.loc[df['Percentage'].idxmax()]
failed = df[df['Grade']=='F']

print("\n Class Topper:")
print(topper[['Name','Percentage','Grade']])

print("\n Failed Students List:")
print(failed[['Name','Percentage','Grade']])
```

##Visualization on Average marks per subject For comparison with Bar Chart

```python
#For the comparison used Bar Chart
print(":-:-:-:-Visualization with Bar chart:-:-:-:-")
plt.figure(figsize=(8,5))
avg_marks = df[subjects].mean()
sns.barplot(x=avg_marks.index, y=avg_marks.values, palette='coolwarm')
plt.title("Average Marks per Subject")
plt.ylabel("Average Marks")
plt.show()
```

##Percentage Distribution

```python
#For distribution used histplot
print(":-:-:-:-Visualization with Histplot:-:-:-:-")
plt.figure(figsize=(8,5))
sns.histplot(df['Percentage'], bins=10, kde=True, color='green')
plt.title("Percentage Distribution")
plt.xlabel("Percentage")
plt.ylabel("Number of Students")
plt.show()
```

##Grade Distribution

```python
#Used countplot
print(":-:-:-:-Visualization with Countplot:-:-:-:-")
plt.figure(figsize=(6,4))
sns.countplot(x='Grade', data=df, palette='magma')
plt.title("Number of Students per Grade")
plt.show()
```

Line chart

```python
#For the showing trend used Line Chart
print(":-:-:-:-Line chart:-:-:-:-")
# Select top 20 students by Percentage
top20 = df.sort_values('Percentage', ascending=False).head(20)

plt.figure(figsize=(14,6))

# Plot line trend
plt.plot(top20['Name'], top20['Percentage'], marker='o', linestyle='-', color='royalblue', linewidth=2)

# Annotate grades above points
for i, r in top20.iterrows():
    plt.text(r['Name'], r['Percentage']+0.5, r['Grade'], ha='center', fontsize=9, color='darkgreen')

plt.title('Top 20 Students Percentage Trend with Grades', fontsize=14, fontweight='bold')
plt.xlabel('Student Name')
plt.ylabel('Percentage')
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Correlation Heatmap

```python
#correlation between subjects
print(":-:-:-:-Correlation Heatmap:-:-:-:-")
plt.figure(figsize=(7,5))
sns.heatmap(df[subjects+['Percentage']].corr(), annot=True, cmap='Blues')
plt.title("Correlation Between Subjects")
plt.show()
```

##Top 5 Student Bar Chart

```python
#Here compare Top 5 Students
print(":-:-:-:-Top 5 Student:-:-:-:-")
plt.figure(figsize=(8,5))
sns.barplot(x='Name', y='Percentage', data=sorted_df.head(5), palette='viridis')
plt.title("Top 5 Students by Percentage")
plt.xticks(rotation=45)
plt.show()
```

##Scatter Plot

```python
#here is scatter plot Maths VS Science Subject
print(":-:-:-:-Scatter Plot:-:-:-:-")
df[['Maths','Science','Percentage']] = df[['Maths','Science','Percentage']].apply(pd.to_numeric)

# Grade colors
colors = {'A':'green','B':'blue','C':'orange','D':'red','F':'purple'}
plt.figure(figsize=(9,5))
for g, c in colors.items():
    sub = df[df['Grade']==g]
    plt.scatter(sub['Maths'], sub['Science'], s=sub['Percentage'], color=c, alpha=0.7, label=g)

plt.title('Maths vs Science Scatter Plot')
plt.xlabel('Maths Marks')
plt.ylabel('Science Marks')
plt.legend(title='Grade')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

##Final Summary

```python
#Here we have final summary of our dataset
print("\n FINAL SUMMARY:\n")
print(f"Total Students: {df.shape[0]}")
print(f"Class Average Percentage: {round(df['Percentage'].mean(),2)}")
pass_percent = round((len(df[df['Grade'] != 'F'])/len(df))*100,2)
print(f"Pass Percentage: {pass_percent}%")
```
